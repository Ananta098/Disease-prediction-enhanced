```python
disease_counts = df['Disease'].value_counts()
plt.figure(figsize=(12, 6))
disease_counts.head(20).plot(kind='bar')
plt.title('Top 20 Diseases in Dataset')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```
